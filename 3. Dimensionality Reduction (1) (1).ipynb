## Dimensionality Reduction

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Urban_Pop_Change.csv', index_col = 0, thousands = ',')
df.index.names = ['Country']
df.columns.names = ['Years']
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(df)
```

```python
transform_data = pca.transform(df)
new_df = pd.DataFrame(transform_data)
new_df.index = df.index
new_df.columns = ['PC1', 'PC2', 'PC3']
new_df.head()
```

```python
print(pca.explained_variance_ratio_)
```
