# Support Vector Machine (SVM) 
### Comparison and effects/ Benefits of Normalization/ Standardization 

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
#Checking the columns
df.columns
```

```python
df.info()
```

```python
#Checking for any null values
df.isnull().sum()
```

```python
#Using One-hot encoding method
dummies = pd.get_dummies(df.diagnosis)
dummies.head()
```

```python
data = pd.concat([dummies, df], axis = 1)
```

```python
data.head()
```

```python
data.drop(['B', 'id', 'diagnosis', 'Unnamed: 32'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
X = data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

y = data['M']  #This is the dependent variable column
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                test_size=0.3, random_state=101)
```

### Performing SVC Before using Standard Scaler

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
model.score(X, y)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

### Performing Standard Scaler to the dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data.drop('M', axis = 1))
```

```python
scaled_feature = scaler.transform(data.drop('M', axis = 1))
```

```python
scaled_data = pd.DataFrame(scaled_feature, columns = data.columns[1:])
```

```python
scaled_data.head()
```

```python
X = scaled_data
y = data['M']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model1 = SVC()
```

```python
model1.fit(X_train, y_train)
```

```python
predictions1 = model1.predict(X_test)
```

```python
model1.score(X, y)
```

```python
print(confusion_matrix(y_test, predictions1))
print('\n')
print(classification_report(y_test, predictions1))
```

### Performing Grid Search

#### Searching for the best perameters 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000, 10000], 'gamma': [1,0.1,0.01,0.001,0.0001, 0.00001], 'kernel': ['rbf']} 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
#The best parameter are
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, grid_predictions))
print('\n')
print(classification_report(y_test, grid_predictions))
```

```python
grid.score(X, y)
```
